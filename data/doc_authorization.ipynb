ИНИЦИАЛИЗАЦИЯ

```python
from users import UserDatabase

db = UserDatabase("postgresql://postgres:111@localhost:5432/authorization")
await db.connect()  # создаёт пул подключений
```

МЕТОДЫ

```python
create_user(username, email, password_hash, role="user") → int

Создаёт пользователя.

Возвращает: сгенерированный id 
Исключения: asyncpg.UniqueViolationError, если email/username уже существуют.



get_user_by_id(id) → dict | None

Возвращает словарь с полями пользователя или None, если не найден (или не активен).


get_user_by_email(email) → dict | None
Аналогично, по email.


get_user_by_username(username) → dict | None
Аналогично, по username.


update_user(id, **updates) → bool

Обновляет указанные поля. Поддерживаются: email, username, role.

Автоматически обновляет updated_at.

Возвращает: True, если запись найдена и обновлена, иначе False.



deactivate_user(id) → bool

Устанавливает is_active = FALSE.
Возвращает: True, если пользователь существовал.


user_exists(email=None, username=None) → bool

Проверяет существование активного пользователя.
Нужно передать либо email, либо username.
```
